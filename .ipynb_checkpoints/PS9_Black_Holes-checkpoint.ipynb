```python
# c. classify the star. get a mass.
def Temp(wvlen, fluxs):
    # wavelength for which flux is max
    wavmax = np.median([wvlens[np.argmax(flux)] for flux in fluxs])
    plt.title('Temperature related to profile-maximizing wavelength')
    plot_all(wvlens,sfluxs)
    plt.axvline(wavmax, linewidth=1, ls='dashed')
    plt.axis([3600,5400,0,1.2])
    # Wein's law
    b = 3.0e+7 # for angstrom kelvin units
    T = b / wavmax
    return T

# T^4 = luminosity
# lum = mass^3/5
```

```python
from scipy.ndimage.interpolation import shift

def wv_diff(test,template,delt=2,steps=1000):
    step = delt / float(steps)
    vectors = np.arange(-delt,delt,step)
    shifts = np.array([shift(test, x) for x in vectors])
    chi2s = np.array([chi_balmer_windows(shift1,template) for shift1 in shifts])
    #         opt_shift = shifts[np.argmin(chi2s)]
    #         compare(balmer[0], template, opt_shift)
    vect_opt = vectors[np.argmin(chi2s)]
#     print vect_opt
    binsize = wvlens[1] - wvlens[0]
    wv_shift = vect_opt * binsize
    return wv_shift

# speed of light in m/s
from scipy.constants import c
# returns in m/s
def redshift_vel(del_lamb, lamb_em):
    return c * (del_lamb / lamb_em)
def delta_vels(template, fluxs):
    max_em = np.max(wvlens)
    return np.array([redshift_vel(wv_diff(test,template), max_em) for test in fluxs])
        
```

```python
def jd_to_sec(period):
    per = Time(period, format='jd')
    zero_pt = Time(0, format = 'jd')
    res = per.cxcsec - zero_pt.cxcsec
    if 3e5 < res and res < 5e5:
        print 'reasonable'
    else:
        print 'unreasonable'
    return res
```
